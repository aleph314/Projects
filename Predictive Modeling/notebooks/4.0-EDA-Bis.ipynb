```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
energy = pd.read_pickle('../data/processed/energy_processed.pkl')
```

Now I'm plotting these new variables to get an idea of how the PUN compares across them:

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.violinplot(x=energy['sun'] + energy['sat'], y=energy['pun'], ax=ax)
ax.set_title('PUN - Weekends vs. Others')
ax.set_xlabel('')
ax.set_ylabel('PUN (€)')
ax.set_xticklabels(['Other Days', 'Weekends']);
# plt.savefig('../reports/figures/violinplot_weekend.png');
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.violinplot(x=energy['holiday'], y=energy['pun'], ax=ax)
ax.set_title('PUN - Holidays vs. Others')
ax.set_xlabel('')
ax.set_ylabel('PUN (€)')
ax.set_xticklabels(['Other Days', 'Holidays']);
# plt.savefig('../reports/figures/violinplot_holiday.png');
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.violinplot(x=(energy['dec'] + energy['jan'] + energy['feb'])*1\
                    + (energy['mar'] + energy['apr'] + energy['may'])*2\
                    + (energy['jun'] + energy['jul'] + energy['aug'])*3\
                    + (energy['sep'] + energy['oct'] + energy['nov'])*4,
               y=energy['pun'],
               ax=ax)
ax.set_title('PUN - Seasons')
ax.set_xlabel('')
ax.set_ylabel('PUN (€)')
ax.set_xticklabels(['Winter', 'Spring', 'Summer', 'Autumn']);
# plt.savefig('../reports/figures/violinplot_season.png');
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.violinplot(x=energy['h8'] + energy['h9'] + energy['h10']\
                    + energy['h11'] + energy['h12'] + energy['h13']\
                    + energy['h14'] + energy['h15'] + energy['h16']\
                    + energy['h17'] + energy['h18'] + energy['h19'] + energy['h20'],
               y=energy['pun'],
               ax=ax)
ax.set_title('PUN - Peakload Hours')
ax.set_xlabel('')
ax.set_ylabel('PUN (€)')
ax.set_xticklabels(['Other Hours', 'Peakload Hours']);
# plt.savefig('../reports/figures/violinplot_evening.png');
```

The PUN seems to be significantly different for all of the new variables.
